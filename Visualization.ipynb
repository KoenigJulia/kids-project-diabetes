### Werte ersetzen mit Median

```python
from sklearn.impute import SimpleImputer
```

```python
simple_imputer = SimpleImputer(strategy='median',missing_values=0)

median_diabetes = diabetes = pd.read_csv('diabetes.csv')

simple_imputer.fit(median_diabetes[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']])

median_diabetes[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = simple_imputer.transform(median_diabetes[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']])

median_diabetes
```

```python
median_diabetes.describe()
```

### Werte mit Durchschnitt ersetzen

```python
from sklearn.impute import SimpleImputer
```

```python
simple_imputer = SimpleImputer(strategy='mean',missing_values=0)\

mean_diabetes = diabetes = pd.read_csv('diabetes.csv')

simple_imputer.fit(mean_diabetes[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']])
mean_diabetes[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = simple_imputer.transform(mean_diabetes[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']])

mean_diabetes
```

```python
mean_diabetes.describe()
```

## Verteilung des Datensatz

```python
diabetes['Outcome'].value_counts()
```

```python
diabetes['Outcome'].value_counts().plot(kind='pie')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X = diabetes[diabetes.columns[:-1]]
y = diabetes['Outcome']

train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=33, stratify=y)

dt_model = DecisionTreeClassifier(random_state=33)
dt_model.fit(train_X, train_y)

pred_y = dt_model.predict(test_X)

accuracy_score = metrics.accuracy_score(test_y, pred_y)
print('Accuracy: {:.2%}'.format(accuracy_score))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(test_y, pred_y)                   
cm_display = ConfusionMatrixDisplay(cm)
cm_display.plot()
```

## Was fällt auf?
Die statistischen Daten haben immer ein Minimum von 0. Aber es ist nicht möglich, dass ein Mensch keine Haut hat oder BMI von 0 hat.
