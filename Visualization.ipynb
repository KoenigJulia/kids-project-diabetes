# Diabetes => Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
diabetes_data = pd.read_csv('diabetes.csv')
diabetes_data
```

```python
diabetes_colors = sns.xkcd_palette(['apple green', 'cranberry'])
sns.palplot(diabetes_colors)
```

```python
sns.boxplot(data=diabetes_data, x='Outcome', y='Glucose', palette=diabetes_colors, whis=[1,99])
```

```python
sns.boxplot(data=diabetes_data, x='Outcome', y='Glucose', color='lightgray', whis=[1,99])
sns.swarmplot(data=diabetes_data, x='Outcome', y='Glucose', palette=diabetes_colors, hue='Outcome')
```

```python
sns.violinplot(data=diabetes_data, x='Outcome', y='Glucose', palette=diabetes_colors, hue='Outcome')
```

# Was hat Insulin mit dem Ergebnis zu tun?

```python
sns.violinplot(data=diabetes_data, x='Outcome', y='Insulin', palette=diabetes_colors, hue='Outcome')
```

```python
sns.histplot(data=diabetes_data, x='Insulin', palette=diabetes_colors, hue='Outcome', bins=10)
```

```python
sns.histplot(data=diabetes_data, x='Insulin', palette=diabetes_colors, hue='Outcome', element='poly')
```

```python
sns.violinplot(data=diabetes_data, x='Outcome', y='Insulin', color='lightgray', hue='Outcome',split='true', inner='quarter')
sns.stripplot(data=diabetes_data, x='Outcome', y='Insulin', palette='cool', hue='Outcome', dodge='true')
```
